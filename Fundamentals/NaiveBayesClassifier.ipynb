```python
import pandas as pd
df = pd.read_csv('titanic.csv')
df.head()
```

```python
#we drop less important features

df.drop(['PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis='columns', inplace=True)
df.head()
```

```python
#separate target variable
target = df.Survived
inputs = df.drop('Survived', axis='columns')
```

```python
#Sex column is text, so we have to ocnvert it into dummies as ML does not handle text
dummies = pd.get_dummies(inputs.Sex)
dummies.head()
```

```python
#add dummies to inputs with concat
inputs = pd.concat([inputs, dummies], axis='columns')
inputs.head()
```

```python
# we don't longer need Sex columns
inputs.drop('Sex', axis='columns', inplace=True)
inputs.head()
```

```python
#Check if there're NaN values
#inputs.columns[inputs.isna().any]
inputs.Age[:10]
```

```python
inputs.Age = inputs.Age.fillna(inputs.Age.mean())
inputs.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs, target, test_size=.2)
```

```python
# we use Gaussian Naive bayes
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train, y_train)

model.score(X_test, y_test)
```
