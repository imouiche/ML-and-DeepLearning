```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
df =pd.read_csv('income.csv')
df.head()
```

```python
plt.scatter(df['Age'], df['Income($)'])
```

```python
#we can assume k=3
km = KMeans(n_clusters=3)
km
```

```python
#assign labels to cluster
y_predicted = km.fit_predict(df[['Age','Income($)']])
y_predicted
```

```python
#plot to see on the graph
#add columns
df['cluster'] = y_predicted
df.head()
```

```python
df0 =df[df.cluster==0]
df1 =df[df.cluster==1]
df2 =df[df.cluster==2]
#scatter plot

plt.scatter(df0.Age, df0['Income($)'], color='green')
plt.scatter(df1.Age, df1['Income($)'], color='red')
plt.scatter(df2.Age, df2['Income($)'], color='black')

plt.legend()
plt.xlabel('Age')
plt.ylabel('Income($)')
```

```python
scaler = MinMaxScaler()
scaler.fit(df[['Income($)']]) #scaller btween 0-1
df['Income($)'] = scaler.transform(df[['Income($)']])
df
```

```python

scaler.fit(df[['Age']])
df.Age = scaler.transform(df[['Age']])
df
```

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[['Age','Income($)']])
y_predicted
```

```python
df['cluster'] =y_predicted
#df.drop('cluster', axis='columns', inplace=True)
df
```

```python
df0 =df[df.cluster==0]
df1 =df[df.cluster==1]
df2 =df[df.cluster==2]
#scatter plot

plt.scatter(df0.Age, df0['Income($)'], color='green', label='Income($)')
plt.scatter(df1.Age, df1['Income($)'], color='red',  label='Income($)')
plt.scatter(df2.Age, df2['Income($)'], color='black',  label='Income($)' )

plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], color='purple', marker='*', label='centroid')
plt.legend()
plt.xlabel('Age')
plt.ylabel('Income($)')
```

```python
km.cluster_centers_
```

```python
#real life data with many features, we can use Algo approach

k_rng = range(1, 10)
sse = [] #some square errors
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df[['Age', 'Income($)']])
    sse.append(km.inertia_)
    
```

```python
sse
```

```python
plt.xlabel('K')
plt.ylabel('Some of squared error')
plt.plot(k_rng, sse)
```
